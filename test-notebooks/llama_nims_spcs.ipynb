```python
import requests
import json
import os
from dotenv import load_dotenv

load_dotenv()
pat = os.getenv("pat")

url = "https://nnc42qqb-sfsenorthamerica-cguzman-aws-us-west-2.snowflakecomputing.app/v1/chat/completions"

headers = {
    "Content-Type": "application/json",
    "Authorization": f"Snowflake Token=\"{pat}\"",
}

payload = {
    "model": "meta/llama-3.1-8b-instruct",
    "messages": [
        {"role": "system", "content": "You are a helpful assistant."},
        {"role": "user", "content": "What is Snowpark Container Services in one sentence?"}
    ],
    "max_tokens": 256,
    "temperature": 0.7
}

response = requests.post(url, headers=headers, json=payload)
result = response.json()
print(result["choices"][0]["message"]["content"])
```
